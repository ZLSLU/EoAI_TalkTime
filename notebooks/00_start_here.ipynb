### Import and configure

We use the `requests` library to talk to the Sproochmaschinn API. The `BASE_URL` points to the server running the speech services.

```python
import requests

BASE_URL = "https://sproochmaschinn.lu"
```

### Create a session

This sends a POST request to create a new session on the server. If you see a session ID printed below, your connection is working and you're ready to go.

```python
session = requests.post(f"{BASE_URL}/api/session").json()
session_id = session["session_id"]
print("Session created:", session_id)
```
