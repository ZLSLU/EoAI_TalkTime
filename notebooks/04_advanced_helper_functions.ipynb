Goal: stop repeating boilerplate and start using reusable functions.

By now you've written the same session → submit → poll pattern several times. The file `sproochmaschinn_helpers.py` wraps all of that into clean functions. Take a look at the file if you're curious how they work.

Import the helper functions. We add the parent directory to the path so Python can find the `sproochmaschinn_helpers` module.

```python
import sys
from pathlib import Path

sys.path.append("..")

from sproochmaschinn_helpers import (
    create_session,
    submit_tts,
    submit_stt,
    poll_result,
    export_result,
    decode_base64_wav_to_file,
)

BASE_URL = "https://sproochmaschinn.lu"
```

### Using the helpers

Compare this to the 5+ cells it took in notebook 01 — same result, much less code.

Create a session with a single function call.

```python
session_id = create_session(BASE_URL)
print("Session:", session_id)
```

Submit a TTS request using `submit_tts` — no need to manually build the POST request ourselves.

```python
request_id = submit_tts(
    base_url=BASE_URL,
    session_id=session_id,
    text="Haut léiere mir wéi een d'Sproochmaschinn benotzt!",
    model="claude",
)
print("TTS request_id:", request_id)
```

`poll_result` handles the polling loop for us, including timeouts and error checking.

```python
result = poll_result(BASE_URL, request_id)
result
```

Decode the base64 audio and save it to a file in one line.

```python
output_path = decode_base64_wav_to_file(result["result"]["data"], "advanced_tts.wav")
print("Saved to:", output_path)
```

Listen to the generated audio.

```python
from IPython.display import Audio

Audio("advanced_tts.wav")
```
